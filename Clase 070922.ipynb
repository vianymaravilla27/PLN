### Text: list of words
### Word: a word to which similarity is measured of all other words in text

```python
def similar_words(text,
                 word,
                 window=8, # Define # palabras de contexto que vamos a tomar 
                 probability=True #Frecuencia normalizada
                 tf_idf=False
                 dot_product=False
                 cosine=True):
    
# contexto_words('dar','estado',..)
# contexto_vector('3,2,10,6')
# El vector de arriba es igual a 21

# Para encontrar el vector de probabilidades es:
# sum(np.array([3,2,10,6]))
# vector_prob = (3/21,2/21,10/21,6/21)
```

```python
# Procesamiento de las listas de las palabras y vectores 
vocabulary = sorted(list(set(text)))
contexts=()
for w in vocabulary:
    context= []
    for i in range(len(text)):
        if text[i]== w: 
            for j in range(i-int(window/2),i): #left context
                if  j >= 0:
                    context.append(text[j])
                    try:
                        for j i range (i+1, i+(int(window/2)+1)) #Right context
                        context.append(text[j])
                        except IndexEror:
                            pass
                        contexts[w]=context
            
```

```python
# Debemos pasarlo luego a numeros 

import numpy as np

if probability:
    vectors = {}
    #counter = 0
    for v in vocabulary:
        context = contexts[v]
        vector = []
        for voc in vocabulary:
            vector.append(context.count(voc))
            vector = np.array(vector)
            s = np.sum(vector)
            vector = vector / s
            vectors[v] = vector
            # counter +=1
            # print(counter)
            
            
            # Calculo con vectores
            # Producto punto
            if dot_product:
                similarities= {}
                v = vectors[word]
                for w in vectors.keys()
                similarities [w] = np.dot(vectors[w],v)
                similarities = (sorted(similarities.items(),
                                      key=lambda item: item[1], reverse= True))
                with open ('similar_words_with_dot_product_of_prob_vectors_to_' + word + '.txt',
                          'w', encoding='UTF-8') as f:
                    for item in similarities:
                        f.write(str(item)+'\n')
                        
                        # Coseno 
                        # El coseno es el producto punto normalizado
                        # El coseno del angulo entre los vectores es el producto punto normalizado
                        if cosine:
                            similarities = {}
                            v1 = vectors[word]
                            from w in vectors.keys():
                                v2 = vectors[w]
                                similarities [w] = np.dot(v1,v2)/(np.linalg.norm(1)*np,linalg,norm(v2))
                                similarities = (sorted(similarities.items(),
                                      key=lambda item: item[1], reverse= True))
                                with open ('similar_words_with_dot_product_of_prob_vectors_to_' + word + '.txt',
                                'w', encoding='UTF-8') as f:
                                    
                                    
                                
                            
                            
                            __________________
                            
                            if _name_=='main'
                            from normalize import normalize
                            from lemmatize import lemmatiza
                            
                            corpus_path=''
                            m=10
                            text= normalize(corpus_path,m)
                            text= lemmatize(text)
                            
                            window = 8
                            word = 'gobierno'
                            similar_words(text, 
                                          word,
                                         window,
                                         probability= True
                                         tf_idf=False
                                         dot_product=False
                                         cosine=True)
                
                
```
